```python
import numpy as np
import pandas as pd
```

```python
raw_configs = pd.read_csv('./regnet_convs.csv')
raw_configs.tail()
```

```python
cfg_str_set = set()
unique_configs = pd.DataFrame()
for i, row in raw_configs.iterrows():
    cfg_str = '_'.join([str(x) for x in row.to_dict().values()])
    if cfg_str not in cfg_str_set:
        cfg_str_set.add(cfg_str)
        unique_configs = pd.concat([unique_configs, raw_configs.iloc[i:i+1, :]], ignore_index=True)
unique_configs.tail()
```

```python
np.random.seed(2022)
unique_configs = unique_configs.sample(frac=1).reset_index(drop=True)
unique_configs.tail()
```

```python
with open('./regnet_convs_unique.csv', 'w') as f:
    f.write(unique_configs.to_csv(index=False))
```
